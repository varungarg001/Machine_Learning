## Working with dates

```python
# converting to datetime datatype

date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

## extract year

```python
date['date_year']=date['date'].dt.year
```

```python
date.head()
```

```python
# Extract months

date['date_month']=date['date'].dt.month
```

```python
date.head()
```

```python
date['date_month_name']=date['date'].dt.month_name()
```

```python
date.head()
```

```python
# Extract days

date['date_days']=date['date'].dt.day
```

```python
date.head()
```

```python
date['date_days_name']=date['date'].dt.day_name()
```

```python
date.head()
```

```python
# is weekend

date['is_weekend']=np.where(date['date_days_name'].isin(['Saturday','Sunday']),1,0)
```

```python
date.head()
```

```python
# Extract week of the year

date['week_of_the_year']=date['date'].dt.isocalendar().week
```

```python
date.head()
```

```python
# Extract quarter

date['quarter_year']=date['date'].dt.quarter
```

```python
date.head()
```

```python
date['semester']=np.where(date['quarter_year'].isin([1,2]),1,2)
```

```python
date.head()
```

```python
# Extract time elasped between dates

import datetime

today=datetime.datetime.today()

today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python
np.round((today-date['date'])/np.timedelta64(30,'D'),0)
```
