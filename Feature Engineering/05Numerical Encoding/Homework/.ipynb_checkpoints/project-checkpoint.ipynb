```python
import pandas as pd
import seaborn as sns
```

```python
from sklearn.preprocessing import Binarizer
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv("Weather_Data.csv",usecols=['MaxTemp','Rainfall'])
```

```python
df.sample(5)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['MaxTemp']
```

```python
# Binarization
trf=ColumnTransformer([('trf1',Binarizer(threshold=1,copy=False),['Rainfall'])],remainder='passthrough')
```

```python
df_transformed=trf.fit_transform(df)
```

```python
df_transformed
```

```python
df.columns
```

```python
df=pd.DataFrame(df_transformed,columns=['RainFall','MaxTemp'])
```

```python
df.sample(5)
```

```python
# Custom Binning


bins=[-5,0,10,20,47]
labels=['Very Cold','Cold','Hot','Very Hot']

df['Weather']=pd.cut(df['MaxTemp'],bins=bins,labels=labels)

```

```python
df.sample(10)
```

```python
df['Weather'].value_counts()
```

```python
df[df['Weather']=='Cold']
```
