```python
import numpy as np
import pandas as pd
```

```python

df = pd.read_csv('titanic.csv')
```

```python
df['number'].unique()
```

```python
# extract numerical column

df['numerical_number']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
df.head()
```

```python
df['number_categorical']=np.where(df['numerical_number'].isnull(),df['number'],np.nan)
```

```python
df.drop(columns=['ticket_cat','ticket_num'],inplace=True)
```

```python
df['ticket_num']=df['Ticket'].str.split(" ").str[-1]
```

```python
df['ticket_num']=pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')
```

```python
df['ticket_cat']=df['Ticket'].str.split(" ").str[0]
```

```python
df['ticket_cat']=np.where(df['ticket_cat'].str.isdigit(),np.nan,df['ticket_cat'])
```
