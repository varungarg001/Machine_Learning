```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("./train.csv")
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
# Step 1 -> train/test/split

X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
df.isnull().sum()
```

```python
# Apply Imputation

si_age=SimpleImputer()
si_embarked=SimpleImputer(strategy='most_frequent')

si_age_train=si_age.fit_transform(X_train[['Age']])
si_embarked_train=si_embarked.fit_transform(X_train[['Embarked']])

si_age_test=si_age.transform(X_test[['Age']])
si_embarked_test=si_embarked.transform(X_test[['Embarked']])
```

```python
# si_age_train
# si_embarked_train
```

```python
# One hot encoding sex and embarked

ohe_sex = OneHotEncoder(handle_unknown='ignore',sparse_output=False)
ohe_embarked = OneHotEncoder(sparse_output=False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(si_embarked_train)

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(si_embarked_test)
```

```python
X_train_embarked
```

```python

X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transormed=np.concatenate((si_age_train,X_train_rem,X_train_sex,X_train_embarked),axis=1)
```

```python
X_test_transfromed=np.concatenate((si_age_test,X_test_rem,X_test_sex,X_test_embarked),axis=1)
```

```python
X_test_transfromed.shape
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train_transormed,y_train)
```

```python
y_pred=clf.predict(X_test_transfromed)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```
