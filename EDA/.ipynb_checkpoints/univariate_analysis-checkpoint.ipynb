```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./train.csv')
```

```python
df.sample(5)
```

# 1. Working with categorical data?

## a. Countplot

```python
sns.countplot(df,x='Survived')
```

```python
# df['Pclass'].value_counts().plot(kind='bar')
sns.countplot(df,x='Pclass')
```

## b. pie chart

```python
values=(df['Survived'].value_counts())
plt.pie(values,labels=['Not survive','survive'],autopct="%1.2f%%")
plt.show()
```

# 1. Working with numerical data

## a. Histogram

```python
plt.hist(df['Age'],bins=10)
plt.show()
```

## b. Distplot

```python
# sns.distplot(df['Age'])

sns.histplot(data=df,x="Age",kde=True)
```

## c. BoxPlot

```python
# 5 number summary

sns.boxplot(df['Age'])
```
